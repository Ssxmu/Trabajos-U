## Estructura selectiva simple 'if' 
La estructura selectiva simple se utiliza para ejecutar un bloque de código solo si una condición es verdadera. Si la condición es falsa, el bloque de código no se ejecuta..“



#### 1. Ejercicio en clase:
Escribe un programa que, dado un número ingreso por teclado, determine si este número es mayor que 10.

```python
#entrada

n = int(input("Introduce un número: "))

#proceso

if n > 10: 
    print(f"El número {n} es mayor que 10")
print("Fin")
```

#### 3. Ejercicio en clase: Distancia entre dos números.
Calcular La distancia que existe entre dos números reales, por otra parte, es el valor absoluto de su diferencia. Entre 8 y 5, por ejemplo, hay una distancia de 3. Esta diferencia tiene un valor absoluto de |3|. Imprima el resultado.

```python

n1 = float(input("Introduce un primer número: "))
n2 = float(input("Introduce un segundo número: "))

#Proceso 

DisN = n1 - n2

if DisN < 0:
    DisN = -DisN
print(f"La distancia entre {n1} y {n2} es: {DisN}")
```

```python
#V2

n1 = float(input("Introduce un primer número: "))
n2 = float(input("Introduce un segundo número: "))

#Proceso 

DisN = abs(n1 - n2)
print(f"La distancia entre {n1} y {n2} es: {DisN}")
```

## Estructura selectiva doble alternativa 'if/else'(compuesta)
La estructura selectiva compuesta se utiliza para ejecutar diferentes bloques de código dependiendo de si la condición es verdadera o falsa. Se presenta una doble alternativa.


#### 4. Ejercicio en clase:

#### 5. Ejercicio: Notas
Escriba un programa que solicite al estudiante las tres notas obtenidas en un curso. La nota definitiva del curso se obtiene  al  promediar las tres notas del estudiante. Tenga en cuenta que el curso se aprueba con una nota mínima de 3.0.
El programa debe imprimir la nota definitiva del curso y en otra línea debe imprimir la aprobación o no del curso.

```python
#Entrada 

Nota1 = float(input("Ingresa la primera nota: "))
Nota2 = float(input("Ingresa la segunda nota: "))
Nota3 = float(input("Ingrese la tercera nota: "))

#proceso 

NotaDef = (Nota1 + Nota2 + Nota3)/3

print(f"\nSu nota definitiva es: {NotaDef:.1f}")

if NotaDef >= 3.0:
    print("Usted aprobó el curso")
else:
    print("Usted no aprobó el curso")
```

#### 7. Ejercicio: Hipotenusa
Crea un programa que permita realizar el cálculo de la hipotenusa de un triángulo, vigilando que ningún cateto debe ser menor o igual a cero. Si se diera el caso, imprimir «Error» por pantalla.

Contexto: El teorema de Pitágoras establece que en un triángulo rectángulo, el cuadrado de la hipotenusa (lado de mayor longitud) es igual a la suma de los cuadrados de los catetos (lados menores del triángulo que conforman el ángulo recto). Si el triángulo rectángulo tiene catetos de longitudes a y b, y la longitud de la hipotenusa es c, entonces:
c^2=a^2+b^2


```python
from math import sqrt

a = float(input("Ingresa el primer cateto del triangulo: "))
b = float(input("Ingresa el otro cateto del triangulo: "))


if a > 0 and b > 0:
    h = sqrt(a**2 + b**2)
    print(f"La hipotenusa del triangulo es: {h}")
else:
    print("Error")

'''
if a <= 0 or b <= 0:
    print("Error")
else:
    h = sqrt(a**2 + b**2)
    print(f"La hipotenusa del triangulo es: {h}")
'''    
```
