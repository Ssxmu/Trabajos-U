Un conjunto es una colección no ordenada de objetos únicos.

* Se usan {} para delimitar los conjuntos y los elementos se separan por comas.
* no puede tener elementos repetidos.
* Los conjuntos pueden ser mutables o inmutables. Para que un conjunto sea inmutable se debe usar la función frozenset().
* Dentro de los conjuntos no puede haber otro tipo de colecciones mutables, es decir, una tupla si puede ser un elemento de un conjunto.
  rror





frozenset() es una función que convierte un iterable (como una lista o conjunto) en un conjunto inmutable. Una vez creado, no puedes modificar sus elementos. Esto es útil cuando necesitas asegurarte de que los datos no cambien, por ejemplo, al usar conjuntos como claves en diccionarios o elementos de otros conjuntos.
Cuando usar frozenset(): 
* Cuando necesitamos conjunto de elementos que no deben cambiar.
* Cuando deseamos tilizar conjuntos como claves en diccionarios.

La clase set ofrece cuatro métodos para eliminar elementos de un conjunto:
* discard(elemento)
* remove(elemento)
* pop() elimina un elemento aleatorio del conjunto.
* clear(): Elimina todos los elementos contenidos en conjunto.
* Los métodos remove(), pop() lanzan la excepción “KeyError” cuando elemento no existe.



```python
# Craer un conjunto vacío
conjunto = set()
print(conjunto)
print(type(conjunto))
```

```python
conjunto.add(20)
conjunto.add("Alfredo")
conjunto.add((1,2))
print(conjunto)

```

```python
A = set()
for i in range(1,50,10):
    A.add(i)
print(A)
```

```python

conjunto = {1, 2, 3, 4, 5}

for elemento in conjunto:
    print(elemento)
```
