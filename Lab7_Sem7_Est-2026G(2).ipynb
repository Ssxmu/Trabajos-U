### While True
Crea un ciclo que se repite indefinidamente mientras la condición sea True (verdadera).
Se detiene si dentro del bloque se encuentra con una instrucción como break.

***Break:*** Instrucción que finaliza la ejecución de un ciclo (for o while), interrumpiéndolo por completo y pasando el control a la siguiente instrucción fuera del bucle.

***Continue:*** Instrucción que interrumpe la iteración actual de un bucle, pero permite que el ciclo continúe con la siguiente iteración, omitiendo el código restante en la iteración en curso.

***break*** detiene completamente el bucle.

***continue*** salta a la siguiente iteración sin salir del bucle.

#### Ejercicio 4. Uso del while True y break
Realiza un programa que calcule el salario promedio de varios empleados. finalice el ciclo cuando el usuario ingrese 0. Use el ciclo while True y la sentencia break. Un solo print.


```python
totalSal = 0 #Variable acumuladora
cuentaSal = 0 #Variable contadora

while True:
    salario = float(input("Ingrese el salario del empleado (0 para finalizar): "))
    if salario == 0:
        print("Fin de ciclo")
        break
    totalSal += salario
    cuentaSal += 1

if cuentaSal > 0:
    promedio = totalSal/cuentaSal
    print(f"{promedio:.2f}")
else:
    print("0.00")
```

#### Ejemplo break y continue
Diapositiva: El programa muestra un conteo regresivo desde 10, no muestra el numero si es par y termina cuando el valor de la variable toma el valor de 2.

```python
var1 = 10
while True:
    var1 = var1 - 1
    if var1 == 2:
        print("El número es 2, break")
        break
    if var1 % 2 == 0:
        print("Continue")
        continue
    print(var1)
print("\nFin del ciclo")
```
