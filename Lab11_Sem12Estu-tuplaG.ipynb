## Tuplas
* Las tuplas son conjuntos ordenados de elementos (números, cadenas, etc).

<!-- * Se usan () para delimitar las tuplas y los elementos se separan por comas. -->

* Las tuplas son inmutables

* Una de las grandes ventajas de las tuplas, al igual que las listas, es que se pueden almacenar tipos de datos distintos



Pueden incluir:
Tipos básicos: int, float, str, bool

Otras colecciones:
* listas (1, [2,3]) 
* otras tuplas (1, (2, 3))
* diccionarios  (1, {"a": 10})
* conjuntos  (1, {2, 3})
* Valores mixtos: (1, "hola", 3.5, True)
  
La tupla no cambia de tamaño ni de posiciones.
Aunque la tupla no se puede modificar, si contiene elementos mutables (como listas), esos elementos sí pueden cambiar.

### Iterar una tupla

```python
print("Iteremos una tupla recorriendo los elementos")
tupla = (3,6,9,12,15)
for i in tupla:
    print(i)
```

```python
print("Iteremos una tupla sobre los indices")
tupla = (3,6,9,12,15)
for i in range(len(tupla)):
    print(tupla[i])
```

## Ejercicios con tuplas

Ejercicio 1.
Crea un programa que defina una tupla con los nombres de 6 asignaturas: asignaturas = ("Matemáticas", "Física", "Programación", "Estadística", "Ética", "Inglés")
Luego, utilizando un ciclo for, recorre la tupla y muestra cada asignatura precedida por su número de orden, comenzando desde 1.
Salida esperada:
1. Matemáticas
2. Lengua
3. Fundamentos
...


```python
# Ejercicio 1: Recorrer una tupla con for
asignaturas = ("Matemáticas", "Física", "Programación", "Estadística", "Ética", "Inglés")
for i in range(len(asignaturas)):
    print(f"{i+1}. {asignaturas[i]}")
```

```python
# Ejercicio 1: Recorrer una tupla con while
asignaturas = ("Matemáticas", "Física", "Programación", "Estadística", "Ética", "Inglés")
i = 0
while i < len(asignaturas):
    print(f"{i+1}. {asignaturas[i]}")
    i += 1
```

```python
numeros = (4, 7, 2, 9, 1, 5)
suma = 0
for i in range(len(numeros)):
    suma += numeros[i]
suma
```
