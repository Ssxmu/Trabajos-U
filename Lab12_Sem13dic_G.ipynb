### Diccionarios


* Un diccionario es una colección de datos que asocia sus elementos en pares(pares de objetos), como clave y valor.
* puede contener distintos tipos de datos, enteros reales, cadenas, listas, tuplas, diccionarios.

### Crear diccionarios

* Los diccionarios se delimitan usando llaves { }, sus elementos son separados por coma, al igual que otras colecciones .
* cada elemento se estructura con su clave y valor, separados con dos puntos {clave:valor}.
* por ejemplo: diccionario = {"color":"rojo"}

* Las claves no están ordenadas y no hay claves repetidas. La única manera de acceder a un valor es a través de su clave.
* En un diccionario de Python, una clave debe ser un dato inmutable (que no cambie).
  
* Por eso, no pueden ser claves los tipos de datos mutables, como:
    * listas (list)
    * diccionarios (dict)
    * conjuntos (set)

* Conclusion: los diccionarios forman un conjunto **desordenado** de correspondencias clave-valor.

### Ejercicio 
Cree un programa en Python que genere un diccionario vacío y luego lo llene utilizando un ciclo for.

Las claves del diccionario deben ser los números del 1 al 5 y los valores deben corresponder al cuadrado de cada número.

Finalmente, muestre en pantalla el diccionario resultante.

```python
cuadrados = {}
for i in range(1,6):
    cuadrados[i] = i ** 2
print(cuadrados)
```
