### Ejercicio 1: Número positivo y par
Escribe un programa que lea un número entero y determine si éste es positivo y además es par.

```python
#Entrada

num = int(input("Ingrese un número: "))

#proceso

if num > 0:
    if num % 2 == 0:
        print(f"El número {num} es positivo y par.")
    else:
        print(f"El número {num} es positivo e impar.")
else:
    print(f"El número {num} es negativo.")
```

#### Ejercicio 2: Categoría edad
Escribe un programa que determine la categoría de edad de una persona basándose en su edad. El programa debe comprobar si la persona es mayor o menor de edad. Si la persona tiene 18 años o más, el programa debe clasificarla como "adulto" o "adulto mayor" según si su edad es mayor o igual a 65 años. Si la persona tiene menos de 18 años, el programa debe indicar que es "menor de edad". El programa debe imprimir el mensaje correspondiente a cada categoría.


```python
edad = int(input("Ingrese la edad de la persona en años: "))

if edad >= 18:
    if edad >= 65:
        print("La persona es 'Adulto Mayor'")
    else:
        print("La persona es 'Adulto'")
else:
    print("La persona es 'Menor de edad'")
```

## Estructura múltiple de alternativas'if/elif/else
Se usa cuando queremos tomar decisiones entre varias opciones posibles dentro de un programa. se evalua en orden jerárquico, cuando encuentra la verdadera se ejecuta y termina.

Con if/elif/else le indicamos al programa, 

"Si esto no es verdadero, intenta esto otro, y si todas las condiciones fallan en ser verdaderas, entonces haz es"..



#### Ejercicio 3: Días de la semana
Diseñar un algoritmo que permita leer un número entero entre 1 y 7 y muetre el correspondiente día de la semanan.


```python
n = int(input("Ingrese un número entre 1 y 7: "))

if n == 1:
    print("Lunes")
elif n == 2:
    print("Martes")
elif n == 3:
    print("Miercoles")
elif n == 4:
    print("Jueves")
elif n == 5:
    print("Viernes")
elif n == 6:
    print("Sábado")
elif n ==7:
    print("Domingo")
else:
    print("Error, fuera de rango")
```
