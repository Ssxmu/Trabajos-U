### Semana 7

El ciclo o bucle es una estructura de control que nos permite repetir un bloque de instrucciones cierta cantidad de veces.

El bloque de instrucciones que se repite se suele llamar cuerpo del bucle y cada repetición se suele llamar iteración o ciclo. 


## Ejercicio 1.
Escribe un programa que imprima los números del 1 al 10 utilizando un bucle for.

```python
for i in range(1, 11, 1):
    print("Número", i)
```

## Ejercicio 2.
Realizar un programa que sume los 5 primeros números (1 al 5).


```python
print("El programa comenzó")
suma = 0

for i in range(1, 6):
    print(i)
    suma = suma + i
    
print("La suma es:", suma)
print("El programa terminó")
```

## Ejercicio 3.
Elabore un programa que solicite al usuario ingrese la tabla de multiplicar que quiere visualizar e imprima la tabla completa.



```python
tabla = int(input("Ingrese el número de la tabla que desea visualizar: "))
        
for n in range(11):
    print(tabla, 'x', n, "=", (tabla*n))
```
